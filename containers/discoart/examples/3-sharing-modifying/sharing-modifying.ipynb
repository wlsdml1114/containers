Let us start with a Tweet from [@hxiao](https://twitter.com/hxiao)

https://twitter.com/hxiao/status/1548420247639638016

He mentions the following disco art to share: `discoart-5caa289076a745da801d2e41cb665851`

**NOTE:**
While the default behavior of `discoart` pip module would save any `DocumentArray` object you create to the cloud for convenience, this configuration is set by default NOT to do so without knowingly opting in.

If you wish to use the very convenient (and public) cloud functionality, you would want to run the following python command in a cell before starting your work:

```python
import os

# Disable Opt-out (so you are effectively "opting in" if you ran this:
os.environ["DISCOART_OPTOUT_CLOUD_BACKUP"]= "0"  # opt-in to cloud backup"] 
```

````python
from docarray import DocumentArray

da = DocumentArray.pull('discoart-5caa289076a745da801d2e41cb665851')

# Let's have a look at the image contents...
da.plot_image_sprites(skip_empty=True, show_index=True, keep_aspect_ratio=True)
````

# Make your own variant

````python
from discoart import create

# Start with DocumentArray loaded in previous cells, and make some parameter changes, as an example.
da_modifed = create(
    init_document=da[0],
    name_docarray = "example-green",
    text_prompts = [
        'A Tarot Card portrait of a beautiful young female face by Alessio Albi. She has perfect white skin like silk. She has green hair and stunning green eyes. She has pretty lips and she wears a dark hood over her hair. green crystal shards illuminated by orange fireflies as background, trending on artstation:10',
        'photorealism,realism,realistic,photograph,hyperreal,painting,3d:-1',
        'text,signature,font,blur,dof,shadow:-3'
    ]
)
````

````python
# Let's have a look at the image contents...
da_modifed.plot_image_sprites(skip_empty=True, show_index=True, keep_aspect_ratio=True)
````
